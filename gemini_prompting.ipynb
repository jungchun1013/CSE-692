# Upload Videos

```python
import time

def upload_video(video_file_name):
  video_file = client.files.upload(file=video_file_name)

  while video_file.state == "PROCESSING":
      print('Waiting for video to be processed.')
      time.sleep(10)
      video_file = client.files.get(name=video_file.name)

  if video_file.state == "FAILED":
    raise ValueError(video_file.state)
  print(f'Video processing complete: ' + video_file.uri)

  return video_file
```

Don't worry if the generated URIs do not allow you to view the data in a browser: https://discuss.ai.google.dev/t/method-doesnt-allow-unregistered-callers-callers-without-established-identity-please-use-api-key-or-other-form-of-api-consumer-identity-to-call-this-api/37579/2

File naming convention: clip_{frame}_{object names}_{event type}

Use the frame number to determine which video is the first in the event sequence


## Define the Task Prompt

```python
def prompt_task2_zero_shot(video_file):
  prompt = """Based on the initial event shown in the following video, predict whether the red ball will subsequently enter the green goal region. Provide your reasoning before answering. Answer "Success" if the ball will successfully enter the goal. Answer "Failure" if the ball will fail to enter the goal.

Answer in the following format:

"Reasoning: {reasoning}
Answer: {answer}"
  """
  response = client.models.generate_content(
      model=model_name,
      contents=[
          prompt,
          video_file
      ]
  )

  return response
```

```python
# get list of clips for task 2
# first for general
# file name convention: clip_{frame}_{object names}_{event type}
root_path = "/content/drive/MyDrive/CSE 692/CSE 692 Team 3/causal_segmentation_with_caption_v2/general"

task2_videos = list()
for directory in tqdm.tqdm(os.listdir(root_path)):
  for subdir in tqdm.tqdm(os.listdir(os.path.join(root_path, directory))):
    frames = dict()
    for video in tqdm.tqdm(os.listdir(os.path.join(root_path, directory, subdir, "videos"))):
      if video.endswith(".mp4") and video != "full_video.mp4":
        _, frame, _, _ = video.split("_")
        frames[int(frame)] = os.path.join(root_path, directory, subdir, "videos", video)

    first_clip = min(frames.items(), key=lambda x: x[0])
    task2_videos.append(first_clip[1])
```

```python
import pandas as pd
from scipy.stats import wilcoxon
```

# Gemini 2.0 Flash Improvement from Task 2.1 to Task 3

```python
# load data from Google Sheet
TASK2_1 = "/content/drive/MyDrive/CSE 692/CSE 692 Team 3/Task2 - Gemini 2.0 Flash.csv"
TASK3 = "/content/drive/MyDrive/CSE 692/CSE 692 Team 3/Task 3 - gemini flash 2.0.csv"
```

```python
task2_1 = pd.read_csv(TASK2_1)
task2_1 = task2_1.dropna(subset="path")
task2_1
```

```python
task3 = pd.read_csv(TASK3)
task3 = task3.dropna(subset="file")
task3
```

```python
task2_1["path"][0]
```

```python
task3["file"][0]
```

```python
task2_1["id"] = task2_1["path"].apply(lambda x: x.split("/")[-3])
task2_1.head()
```

```python
task3["id"] = task3["file"].apply(lambda x: x.split("/")[-3])
task3.head()
```

```python
task2_1 = task2_1[["id", "Answer correct? (1 = Correct, 0 = Incorrect)"]]
task2_1.rename(columns={"Answer correct? (1 = Correct, 0 = Incorrect)": "task_2_1_correct"}, inplace=True)
task2_1
```

```python
task3 = task3[["id", "correct"]]
task3.rename(columns={"correct": "task_3_correct"}, inplace=True)
task3
```

```python
# merge dataframes
df = pd.merge(task2_1, task3, on="id")
df
```

```python
# https://www.maartengrootendorst.com/blog/validate/
result = wilcoxon(df["task_2_1_correct"], df["task_3_correct"], zero_method='zsplit')
print("V = ", result.statistic)
print("p = ", result.pvalue)
```

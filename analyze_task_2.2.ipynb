# Analyze LLAVA Results

```python
llava_df = pd.read_csv(LLAVA_RESULTS)
```

```python
llava_df["event_type"] = llava_df["File"].apply(lambda x: x.split("_")[-1].replace(".mp4", ""))
```

```python
llava_df["correct"] = llava_df["Groundtruth"] == llava_df["Prediction"]
```

```python
llava_df.head()
```

```python
# compute accuracy by event type
llava_df.groupby("event_type")["correct"].mean()
```
